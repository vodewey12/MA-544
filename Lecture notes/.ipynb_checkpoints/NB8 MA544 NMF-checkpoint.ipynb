### NMF: Introduction
---
Given a nonnegative matrix $V \in \mathbb{R}^{m,n}$, the process of finding nonnegative matrices $W \in \mathbb{R}^{m,r}$ and $ H \in \mathbb{R}^{r,n}$, where $r \leq \min{(m,n)}$, such that
 
 $$\Large \color{blue}{ \displaystyle V \approx WH}, $$
 
 i.e.  <font color=‘red’>when $WH$ best approximates $V$ in some sense</font>, is called  *nonnegative matrix factorization*. 
 
 In practice, $\color{blue}{r << min(m,\,n)}$ . This leads to an efficient method for <font color=‘blue’>dimensionality reduction of nonnegative data</font>. 
 
 The NMF method was first proposed by Paatero and Tapper. However, it received wide recognition only after Lee and Seung proposed their seminal algorithm and demonstrated its usefulness in part based learning.

### Properties  and Interpretations
---
Images obtained from satellites and several other sources are usually received in multiplicity and  presenting them  in  compact and efficient form is an important task. An image -- which generally could be understood as a matrix of gray scale values of the pixels -- could be stored as a vector by stacking the columns of the image-matrix. Suppose the columns of the matrix $V \in \mathbb{R}^{m \times n}$ represent images, then it is possible to represent an image as a linear combination of some image parts, represented by columns of matrix $W$, as 
<br>

$$\large \color{blue}{ 
\rm{image}_j = (\rm{part}_1)(\rm{weight}_1) + (\rm{part}_2)(\rm{weight}_2) + \dots + (\rm{part}_r)(\rm{weight}_r),
}
$$

<br>

<img src="./images/NMF_VWH.png" width="80%" />


which could be written mathematically as
<br>

$$\Large \color{blue}{
  V_j \approx  W_1 H_{1,j}+ \cdots + W_r H_{r,j}  = \sum_{i = 1}^r  W_i H_{i,j} = WH_{j} ,
  }
$$


  where $V_j$ represents the $j$th column of $V$ and $H_{i,j}$ represents the ${(i,j)}$-th element of the matrix $H$.
  
  
  
  In statistics, finding basis random variables from a set of observed data is of great significance. Suppose that elements of matrix $V$ are data from a set of statistical observations.
$ V_{i,j} = $ Observation $i$ of vector $j$ of values of some random variables.\\
 In this case, the NMF could be interpreted as representations of an element, say $V_{i,j}$ as the linear combination of some  <font color=‘blue’> basis random variables (BRV)</font> given by the row-elements of matrix $W$ as:
 
 <br>
 
$$ \large \color{blue}{
\rm{value}_{i,j} = (BRV_1)(\rm{weight}_1) + (BRV_2)(\rm{weight}_2) + \cdots + (BRV_r)(\rm{weight}_r),
}
$$
which could be expressed as
$$\Large \color{blue}{
V_{i,j} \approx \sum_{k=1}^r  W_{i,k}H_{k,j}.
}
$$
Physically, this means finding  basis random variables from a set of observed data. This gives the latent structure hidden in the data-set and has been shown to be a useful technique in many applications. For instance, in environmetrics it is used to find basic components of pollutants in a set of observations on pollution under varying conditions.

### NMF: Problem Formulation
---
It is standard practice to formulate the NMF problem as an optimization problem which minimizes some type of distance function between $V$ and $WH$. A natural choice of such a function is the Frobenius norm, which results in the following constrained optimization problem:

**NMF Problem Statement**

Suppose  that $V \in \mathbb{R}^{m,n}$ is a nonnegative matrix and $r$ is a positive integer satisfying $ r \le \min\{m,n\}$.

<div class='eqnbox3'>
Find matrices $W$ and $H$ which solve
$$\large \color{blue}{
 \min f(W,H) :=\frac{1}{2} \|V-WH\|_F^2, \ \ \textrm{ s.t. } \ \ W \geq 0, \ H \geq 0,
}$$

where $W \in \mathbb{R}^{m,r}$, $H \in \mathbb{R}^{r,n}$.
</div>

**Uniqueness of NMF**

We notice from the observation $WH = WDD^{-1}H$ that an NMF problem cannot have a unique solution. For any invertible nonnegative matrix $D$ whose inverse is also nonnegative, if $(W,H)$ is a solution, then so is $(WD, D^{-1}H)$ and both give the same approximation. The following lemma is well known (see lemma 1.1 in \cite{Minc}) .

<div class="alert alert-info">
<strong> Lemma: </strong> The inverse of a nonnegative matrix is nonnegative if and only if it is a scaled permutation.
</div>








**EXAMPLE**: The NMF problem above is convex in both $W$ and $H$ separately.


#### Karush-Kuhn-Tucker (KKT)  Conditions for Stationary Points of NMF
---
Following expression is from the WikiPedia link on [KKT Conditions](https://en.wikipedia.org/wiki/Karush%E2%80%93Kuhn%E2%80%93Tucker_conditions#Necessary_conditions)

Suppose we are working on solving the following multivariable (${\bf x} \in \mathbb{R}^n$)
**Optimization Problem:**

Optimize  $f({\bf x})$ 

subject to the inequlaity constrains $ g_i({\bf x}) \le  0 \quad (i=1,\cdots, m)$ 

and equality constraints $ h_j({\bf x})=0 \quad (j=1,\cdots,l ) $.

<br>

 Let $\mathbf g(x) : \,\!\mathbb{R}^n \rightarrow \mathbb{R}^m$ be defined as $\mathbf{g}(x) = \left( g_{1}(x), \ldots, g_{m}(x) \right)^\top$and let $\mathbf h(x) : \,\!\mathbb{R}^n \rightarrow \mathbb{R}^{\ell}$be defined as $\mathbf{h}(x) = \left( h_{1}(x), \ldots, h_{l}(x) \right)^\top$.
 
 Let $\boldsymbol{\mu} = \left( \mu_1, \ldots, \mu_m \right)^\top$ and $\boldsymbol{\lambda} = \left( \lambda_1, \ldots, \lambda_{l} \right)^\top$. 

Then the necessary conditions can be written as:

**Stationarity**
- For maximizing $f(x)$: $$\nabla f(x^*) - D \mathbf g(x^*)^{\top} \boldsymbol{\mu} - D \mathbf h(x^*)^{\top} \boldsymbol{\lambda} = \mathbf 0$$
- For minimizing $f(x)$: $$\nabla f(x^*) + D \mathbf g(x^*)^{\top} \boldsymbol{\mu} + D \mathbf h(x^*)^{\top} \boldsymbol{\lambda} = \mathbf 0$$

**Primal feasibility**
$$\mathbf g(x^*) \le \mathbf 0$$
$$\mathbf h(x^*) = \mathbf 0$$

**Dual feasibility**
$$\boldsymbol \mu \ge \mathbf 0$$

**Complementary slackness**
$$\boldsymbol \mu^{\top} \mathbf g (x^*) = 0.$$

---
**Example**: Find the gradients of $f(W,\,H)$, namely, $\nabla_H f$ and  $\nabla_W f$. (In class)
$$
\nabla_H f = -W^T(V-WH), \text{ and } \nabla_H f = -(V-WH)H^T
$$


**KKT Conditions for NMF**:If W and H solve the NMF problem as given above then there exist KKT multiplier matrices $\mu \geq 0$ and $\nu \geq 0$ such that
<div class='eqnbox2'>
\begin{align}
-(V-WH)H^T &= \mu,\\
-W^T(V-WH) &= \nu,\\
\mu \circ W &= 0,\quad \nu \circ H = 0,\\
    \quad W &\geq 0,\quad H \geq 0
\end{align}
</div>
**Example**: Which of the following statements do you agree with?

>- (A) $-(V-WH)H^T \circ W = 0_{m,r}$

>- (B) $-(V-WH)H^T \circ H = 0_{m,r}$

>- (C) $-W^T(V-WH) \circ H = 0_{r,n}$

>- (D) $-W^T(V-WH) \circ W = 0_{r,n}$


**Example**: Which of the following statements do you agree with
>- (A) $\min \{WHH^T - VH^T, H \} = 0_{m,r}$

>- (B) $\min \{WHH^T - VH^T, W \} = 0_{m,r}$

>- (C) $\min \{W^T WH - W^T V, W \} = 0_{r,n}$

>- (D) $\min \{W^T WH - W^T V, H \} = 0_{r,n}$

**Example**: Prove that the following holds for the stationary points $(W,\,H)$ of NMF.
$$
\|V-WH\|^2_F = \|V\|^2_F - \|WH\|^2_F.
$$


### Algorithms for NMF
---

#### Lee--Seung Algorithm 
[Refer to this article](https://papers.nips.cc/paper/1861-algorithms-for-non-negative-matrix-factorization.pdf)

In a standard gradient descent scheme for NMF such as 

$$\large
H_{ij} \leftarrow H_{ij} + \eta_{ij} \left[(W^TV)_{ij} - (W^TWH)_{ij} \right]
$$

choosing small values of $\eta_{ij}$ ensures reduction in $\|V-WH\|_F$. The following choice leads to a multiplicative algorithm given below.
<div class='eqnbox2'>
$$\large
\eta_{ij} = \frac{H_{ij}}{(W^T WH)_{ij}}
$$
</div>
A number of algorithms have been proposed for solving the NMF problem.  One of the simplest and most discussed algorithms was proposed by  Lee and Seung \cite{LS97}. This is a multiplicative algorithm that starts with a positive pair $(W^0,H^0)$ and iteratively generates a sequence of approximations $(W^k, H^k)$ to a solution of problem \reff{nnmf} using the following updates:
<div class='eqnbox3'>
$$\large \color{blue}{
   H^{k+1}=H^k \ \circ \ ((W^k)^TV) \ ./ \ ((W^k)^TW^kH^k)
}$$
and
$$\large \color{blue}{
  W^{k+1} = W^{k}\ \circ \ (V(H^{k+1})^T)\ ./ \ ((W^k H^{k+1}(H^{k+1})^T)
}$$
</div>
where '$\circ $' is the <font color="green"> elementwise multiplication and '$./$' is the elementwise division operation </font>.


The Lee--Seung  algorithm generates a sequence of approximation pairs $(W^k,H^k)$ by updating $W$ and $H$ alternatively as follows:
<br>

1. **Initialization**: Choose matrices $W^0\in \mathbb{R}^{m,r}_+$ and $H^0\in \mathbb{R}^{r,n}_+$, and set $k=0$. 

1. **Iteration**: Iterate until convergence:

> (1) Update $H$, with $W$ fixed:
        
$$\large \color{blue}{
   H^{k+1}=H^k \ \circ \ ((W^k)^TV) \ ./ \ ((W^k)^TW^kH^k)
}$$
> (2) Update $W$, with $H$ fixed:
        
$$\large \color{blue}{
  W^{k+1} = W^{k}\ \circ \ (V(H^{k+1})^T)\ ./ \ ((W^k H^{k+1}(H^{k+1})^T)
}$$
> (3) Set $k=k+1$.

We refer to the above as the Lee--Seung algorithm. This algorithm  is very easy to implement and has been used in some applications. It has also been extended to nonnegative matrix factorization problems with additional constraints (see \cite{Hoyer04,Li}). Lee and Seung prove the successive nonincrement of the objective function under the suggested updates, i.e.,

$$\large \color{blue}{
f(W^{k+1},H^{k+1})  \leq f(W^{k+1},H^{k}) \leq f(W^{k},H^{k}).
}$$

However, convergence to a KKT-point has not been proved. Under certain conditions, Lin \cite{Lin2} proves that a modification of the Lee-Seung algorithm converges to a KKT point of 
the problem \reff{nnmf}.  It has been found that the Lee--Seung algorithm can be slow (see \cite{Berry,Lin,Cichocki}). In chapter 4, we analyze an algorithm called the {\it Image Space Reconstruction Algorithm} (ISRA) and provide a heuristic explanation for the slow rate of convergence of the Lee--Seung algorithm.

#### Topic Modeling: discovering abstract topics from a collection of texts.
---

Please see term-frequency and inverse document frequency [TF-IDF](https://en.wikipedia.org/wiki/Tf%E2%80%93idf)
<div class='eqnbox2'>
$$
\mathrm{tfidf}(t,d,D) = \mathrm{tf}(t,d) \cdot \mathrm{idf}(t, D)
$$
where
$$
  \quad \mathrm{idf}(t, D) =  \log \frac{N}{|\{d \in D: t \in d\}|}
$$
and term-frequency, tf$(t,d)$, is the relative frequency of the term in a given document.
</div>

This example is a [reproduction of this.](https://nbviewer.jupyter.org/github/fastai/numerical-linear-algebra/blob/master/nbs/2.%20Topic%20Modeling%20with%20NMF%20and%20SVD.ipynb)

For a basic introduction of topic modeling, read this [blog](https://www.toptal.com/python/topic-modeling-python) by Federico Albanese.

**Know about 20newsgroups dataset** [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_20newsgroups.html).

```python
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn import decomposition
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from scipy import linalg
import matplotlib.pyplot as plt

%matplotlib inline
np.set_printoptions(suppress=True)
```

```python
categories = ['alt.atheism', 'talk.religion.misc', 'comp.graphics', 'sci.space']
discard = ('headers', 'footers', 'quotes')
newsgroups_train = fetch_20newsgroups(subset='train', categories=categories, remove=discard)
newsgroups_test = fetch_20newsgroups(subset='test', categories=categories, remove=discard)
```

```python
num_top_words=8

def show_topics(a):
    top_words = lambda t: [vocab[i] for i in np.argsort(t)[:-num_top_words-1:-1]]
    topic_words = ([top_words(t) for t in a])
    return [' '.join(t) for t in topic_words]
```

#### TF-IDF (Term Frequency - Inverse Document Frequency)

```python
vectorizer_tfidf = TfidfVectorizer(stop_words='english')
vectors_tfidf = vectorizer_tfidf.fit_transform(newsgroups_train.data) # (documents, vocab)
```

```python
m,n=vectors.shape
d=5  # num topics
```

```python
clf = decomposition.NMF(n_components=d, random_state=1)

W1 = clf.fit_transform(vectors)
H1 = clf.components_
```

```python
show_topics(H1)
```
